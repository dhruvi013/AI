```python
import pdfplumber

def extract_text_from_pdf(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() + "\n"
    return text

# Example usage:
pdf_files = ["/content/RP-1.pdf", "/content/RP-2.pdf","/content/RP-3.pdf","/content/RP-4.pdf","/content/RP-5.pdf",
             "/content/RP-6.pdf","/content/RP-7.pdf","/content/RP-8.pdf","/content/RP-9.pdf","/content/RP-10.pdf"]
documents = [extract_text_from_pdf(pdf) for pdf in pdf_files]
```

```python
from summa import keywords

def extract_keyphrases(text):
    keyphrases = keywords.keywords(text, scores=True)
    return keyphrases

# Apply to all papers
keyphrases_per_paper = [extract_keyphrases(doc) for doc in documents]
```

```python
def categorize_keyphrases(keyphrases):
    categorized = {"Relevant": [], "Moderately Relevant": [], "Irrelevant": []}

    for phrase, score in keyphrases:
        if "virtual try-on" in phrase or "clothing" in phrase or "3D AR" in phrase:
            categorized["Relevant"].append((phrase, score))
        elif "consumer behavior" in phrase or "image processing" in phrase:
            categorized["Moderately Relevant"].append((phrase, score))
        else:
            categorized["Irrelevant"].append((phrase, score))

    return categorized

# Apply to all papers
categorized_results = [categorize_keyphrases(phrases) for phrases in keyphrases_per_paper]
```

```python
import json

with open("keyphrases_results.json", "w") as file:
    json.dump(categorized_results, file, indent=4)
```

```python
import pandas as pd

df = pd.DataFrame(categorized_results)
print(df)
```
